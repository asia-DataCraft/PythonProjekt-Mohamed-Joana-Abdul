```python
import pandas as pd
import calendar


xlsx_file_path = "Daten_Aufgabe_1 - Shopauswertung.csv"

data = pd.read_csv(xlsx_file_path, delimiter=';')

data.head(10)
```

```python

data["Datum"] = pd.to_datetime(data["Datum"], format='%d.%m.%Y')

data["Monat"] = data["Datum"].dt.month

aggregated_data = pd.DataFrame(columns=["Monat", "Produkt", "Umsatz"])

# Liste der Produkte
produkte = ["A", "B", "A+B+C"]

for monat in data["Monat"].unique():
    for produkt in produkte:
        umsatz = data[(data["Monat"] == monat) & (data["Produkt"] == produkt)]["Umsatz"].sum()
        aggregated_data = aggregated_data.append({'Monat': monat, 'Produkt': produkt, 'Umsatz': umsatz}, ignore_index=True)

aggregated_data
```

```python


produkte = ["A", "B", "C", "A+B+C"]


aggregated_data = {"Monat": [], "Produkt_A": [], "Produkt_B": [], "Produkt_C": [], "A+B+C": []}


for monat in data["Monat"].unique():
    monatliche_daten = data[data["Monat"] == monat]
    umsatz_a = round(monatliche_daten[monatliche_daten["Produkt"] == "A"]["Umsatz"].sum(), 2)
    umsatz_b = round(monatliche_daten[monatliche_daten["Produkt"] == 'B']["Umsatz"].sum(), 2)
    umsatz_c = round(monatliche_daten[monatliche_daten["Produkt"] == "A+B+C"]["Umsatz"].sum() - (umsatz_a + umsatz_b), 2)
    umsatz_abc = round(monatliche_daten[monatliche_daten["Produkt"] == "A+B+C"]["Umsatz"].sum(), 2)

    aggregated_data["Monat"].append(calendar.month_name[monat])
    aggregated_data["Produkt_A"].append(umsatz_a)
    aggregated_data["Produkt_B"].append(umsatz_b)
    aggregated_data["Produkt_C"].append(umsatz_c)
    aggregated_data["A+B+C"].append(umsatz_abc)

df_eu = pd.DataFrame(aggregated_data)

df_eu.style.format(thousands=",")
```
